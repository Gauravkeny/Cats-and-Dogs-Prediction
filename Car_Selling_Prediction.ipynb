```python
data['engine'] = data['engine'].str.replace(' CC','')
data['max_power'] = data['max_power'].str.replace(' bhp','')
data['mileage'] = data['mileage'].str.replace(' kmpl','')
data['mileage'] = data['mileage'].str.replace(' km/kg','')

data.head(10)
```

```python
data['engine'] = pd.to_numeric(data['engine'],errors='coerce')
data['mileage'] = pd.to_numeric(data['mileage'],errors='coerce')
data['max_power'] = pd.to_numeric(data['max_power'],errors='coerce')
```

```python
data.loc[data['max_power'].isnull(),'max_power'] = data['max_power'].median()
data.loc[data['engine'].isnull(),'engine'] = data['engine'].median()
data.loc[data['seats'].isnull(),'seats'] = data['seats'].median()
data.loc[data['mileage'].isnull(),'mileage'] = data['mileage'].median()

data.isnull().sum()
```

```python
data.dtypes
```

```python
X = data.drop(['selling_price'],axis=1)
Y = data['selling_price']
Y = np.log(Y)
```

```python
from sklearn.linear_model import Lasso
```

```python
model = Lasso(fit_intercept=True)
```

```python
from sklearn.model_selection import GridSearchCV,cross_val_score
```

```python
params = {'alpha':np.linspace(0.001,0.01,200)}
```

```python
Grid_lasso = GridSearchCV(model,cv=10,param_grid=params,n_jobs=-1,verbose=1000,scoring='neg_mean_absolute_error')
```

```python
Grid_lasso.fit(X,Y)
```

```python
Grid_lasso.best_estimator_
```

```python
Lasso_last = Grid_lasso.best_estimator_
```

```python
Lasso_last.fit(X,Y)
```

```python
list(zip(X.columns,Lasso_last.coef_))
```

```python
Lasso_cv = cross_val_score(Lasso_last,X,Y,cv=10,n_jobs=-1,verbose=10000,scoring='neg_mean_absolute_error')
```

```python
Lasso_cv.mean()
```
